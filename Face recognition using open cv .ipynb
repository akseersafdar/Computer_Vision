```python
#making the import
import cv2
import matplotlib.pyplot as plt
```

```python
#reading the file in color mode
img = cv2.imread('brad 1.jpg',1)
```

```python
print(img)
```

```python
#checking the shape of numpy array as all images are read into numpy arrays
img.shape
```

```python
plt.figure(figsize = (10,8))
plt.imshow(img)
plt.xticks([])
plt.yticks([])
```

```python
#reading the opencv face detector file (xml format)
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
#draw the face rectangle

grey_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(grey_image, scaleFactor = 1.05, minNeighbors = 5)

print(faces)

for x,y,w,h in faces:
    img = cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 3)
    
    
plt.figure(figsize = (10,8))
plt.imshow(img)
plt.xticks([])
plt.yticks([])
plt.show()
```
